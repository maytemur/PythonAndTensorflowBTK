## instance & attribute

```python
class SuperHeros():

    ozelGuc = "Gorunmezlik"
    
    def __init__(self,isimInput,yasInput,meslekInput):
        print("init çağrıldı")
        self.isim = isimInput
        self.yas = yasInput
        self.meslek = meslekInput
    def ornekMethod(self):
        print (f"ben süperkahramanım ve mesleğim: {self.meslek}")
```

```python
superman = SuperHeros("Superman",35,"Gazeteci")
```

```python
batman = SuperHeros("Batman",43,"İş adamı")
```

```python
superman.isim ="Clark Kent"
```

```python
superman.isim
```

```python
superman.ornekMethod()
```

```python
superman.ozelGuc
```

```python
class Kopek():
    yilCarpani = 7
    def __init__(self,yas=5):
        self.yas = yas
    def insanYasiniHesapla(self):
        return self.yas * self.yilCarpani
```

## Inheritance

```python
class Hayvan():
    def __init__(self):
        print ("hayvan sınıfı init çağrıldı")

    def method1(self):
        print ("hayvan sınıfı method1 çağrıldı")

    def method2(self):
        print ("hayvan sınıfı method2 çağrıldı")
```

```python
benimhayvan = Hayvan()
```

```python
benimhayvan.method1()
```

```python
benimhayvan.method2()
```

```python
class Kedi(Hayvan):
    def __init__(self):
        Hayvan.__init__(self)
        print("kedi sınıfı init çağrıldı")
    def Miyavla(self):
        print ("miyav")    
```

```python
benimKedi = Kedi()
```

```python
benimKedi.Miyavla()
```

```python
benimKedi.method2()
```

## Polymorphism - aynı isimde metotların farklı amaçlara hizmet edebilmesine diyoruz

```python
class Elma():
    def __init__(self,isim):
        self.isim = isim

    def bilgiVer(self):
        return self.isim + " 100 kaloridir "
```

```python
class Muz():
    def __init__(self,isim):
        self.isim = isim

    def bilgiVer(self):
        return self.isim + " 150 kaloridir "
```

```python
elma = Elma("elma")
```

```python
elma.bilgiVer()
```

```python
muz = Muz("muz")
```

```python
muz.bilgiVer()
```

```python
meyveListesi = [elma,muz]
```

```python
for meyve in meyveListesi:
    print(meyve.bilgiVer())
```

```python
def bilgiAl(meyve):
    print (meyve.bilgiVer())
```

```python
bilgiAl(muz)
```

```python
bilgiAl(elma)
```
