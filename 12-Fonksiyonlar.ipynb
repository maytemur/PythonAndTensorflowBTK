## args & kwargs - argümanlar ve key argümanlar

```python
# kullanıcı ne kadar isterse o kadar değişken girebilsin mesela kaç sayı istiyorsa o kadar sayı girsin toplayalım
```

```python
def superToplama (*args):
    return sum(args)
```

```python
superToplama (34,56,78,34,56,342,34)
```

```python
def ornekFonksiyon (**kwargs):
    return(kwargs)
```

```python
type(ornekFonksiyon(muz= 123,elma=34,ananas=345))
```

```python
def keyWordKontrolu (**kwargs):
    if "maytemur" in kwargs:
        print ("maytemur var")
    else:
        print("maytemur yok")
```

```python
keyWordKontrolu (isim=23,mike =2345,mayt=234)
```

```python
keyWordKontrolu (isim=23,maytemur =2345,mayt=234)
```
