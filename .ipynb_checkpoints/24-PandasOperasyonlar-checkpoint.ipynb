## groupby

```python
maasSozlugu = {"Departman" : ["Yazılım","Yazılım","Pazarlama","Pazarlama","Hukuk","Hukuk"],
              "Çalışşan İsmi" : ["Ahmet","Mehmet","Atil","Burak","Zeynep","Fatma"],
              "Maas" : [100,150,235,340,560,670]}
```

```python
maasDataFrame = pd.DataFrame(maasSozlugu)
```

```python
maasDataFrame
```

```python
grupObjesi = maasDataFrame.groupby("Departman")
```

```python
grupObjesi.count()
```

```python
grupObjesi.max()
```

```python
grupObjesi.min()
```

```python
grupObjesi.describe()
```
