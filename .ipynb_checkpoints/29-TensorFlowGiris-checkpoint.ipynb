## Veriyi test/train olarak ikiye ayrmak

```python
from sklearn.model_selection import train_test_split #train_test_split dkmantasyonuna bir bakn dedi
```

```python
# y = wx + b
# y -> label
y=dataFrame["Fiyat"].values # numpy dizisine evirdik

# x -> feature (zellik)
x = dataFrame[["BisikletOzellik1","BisikletOzellik2"]].values
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.33,random_state=15) #bununla artk bldk
```

```python
y
```

```python
x
```

```python
y_test.shape
```

```python
#modelini oluturmadan nce scale etmemiz gerekiyor
#burda code ile yaptmz ileri playground.tensorflow.org zerinde grafik olarakda yapabiliyoruz
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
x_train
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(5,activation="relu"))
model.add(Dense(5,activation="relu"))
model.add(Dense(5,activation="relu"))

model.add(Dense(1))

model.compile(optimizer = "rmsprop",loss="mse") #mse regression denilerek google aratlp teferruatl baklabilir
```

```python
model.fit(x_train,y_train,epochs=250)
```

```python
loss=model.history.history["loss"]
```

```python
sbn.lineplot(x=range(len(loss)),y=loss)
```

```python
trainLoss = model.evaluate(x_train,y_train, verbose=0)
```

```python
model.evaluate(x_train,y_train, verbose=0)
```

```python
testLoss= model.evaluate(x_test,y_test, verbose=0)
```

```python
testLoss
```

```python
testTahminleri = model.predict(x_test)
```
