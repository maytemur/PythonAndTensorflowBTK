```python
benimListe = [12,14,16,18,20]
```

```python
for x in benimListe :
    print(x/5)
```

Belli bir değerde bitmesini istiyorsak break kullanıyoruz

```python
for x in benimListe :
    if x == 16:
        break # bu değeri bulduğunda döngüyü bitir
    print(x/5)
```

Belli bir değeri atlamak için ise continue kullanıyoruz

```python
for x in benimListe :
    if x == 16:
        continue # bu değeri bulduğunda onu atla devam et
    print(x/5)
```

pass ise döngü hata vermesin ama pas geçsin diye kullanıyoruz yani döngü sonra doldurulacaksa falan 

```python
for x in benimListe :
    pass
```
