```python
benimYaslar = [45,46,57]
```

```python
benimIsimler = ["Mustafa", "Zeynep", "Mehmet"]
```

```python
pd.Series(benimYaslar)
```

```python
pd.Series (data=benimYaslar, index = benimIsimler)
```
