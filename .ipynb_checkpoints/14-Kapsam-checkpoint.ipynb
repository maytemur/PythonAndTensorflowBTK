## Local, Enclosing, Global, Built-In

```python
benimAdim = "Mayt"
#Global

def benimFonk():
    benimAdim = "Ayte"
    #Enclosing
    def icFonk():
        benimAdim ="Temur"
        #Local
        print(benimAdim)
    icFonk()
```

```python
benimFonk()
```

```python
print(benimAdim)
```

```python
y=10

def yeniFOnk(y):
    print(y)
    y=5
    print(y)
    return y
```

```python
yeniFOnk(3)
```

```python
y = yeniFOnk(3)
```

```python
y
```

```python
y=10

def ornekFonk():
    global y
    y=5
    print(y)
```

```python
ornekFonk()
```
