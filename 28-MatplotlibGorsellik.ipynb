```python
import numpy as np
import matplotlib.pyplot as pyt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 **2
```

```python
(benimFigur,benimEksen) = pyt.subplots(dpi=150)
benimEksen.plot(numpyDizisi1,numpyDizisi2,"g")
benimEksen.plot(numpyDizisi2,numpyDizisi1,color="#C96F23",alpha=0.3)
```

```python
(yeniFigur,yeniEksen) = pyt.subplots(dpi=150)
yeniEksen.plot(numpyDizisi1,numpyDizisi1+2,"g",linewidth=5.0)
yeniEksen.plot(numpyDizisi1,numpyDizisi1+3,color="yellow")

yeniEksen.plot(numpyDizisi1,numpyDizisi1+4,"r",linestyle="-.")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+6,"b",linestyle=":",marker="o",markersize=8,markerfacecolor="red")
```
