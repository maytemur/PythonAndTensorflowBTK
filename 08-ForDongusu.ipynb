```python
benimListem = [1,2,3,4,5,6]
```

```python
print("döngü başladı")
for numara in benimListem:
    print(numara)
print("döngü bitti")
```

```python
print("döngü başladı")
for numara in benimListem:
    print(numara *2 +2)
print("döngü bitti")
```

```python
print("döngü başladı")
for numara in benimListem:
    if numara %2 ==0: #eğer sayı çiftse
        print("çift numara:",numara)
print("döngü bitti")
```
