```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
yasListesi = [12,23,34,45,57,43,64,34,45,23]
```

```python
kiloListesi = [25,35,46,23,34,56,67,34,43,22]
```

```python
numpyYasListesi = np.array(yasListesi)
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyKiloListesi
```

```python
numpyYasListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"b")
```

```python
plt.plot(numpyKiloListesi,numpyYasListesi,"r")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo'nun Yaşa Göre Değişim Grafiği")
#plt.show()
```

```python
numpyDizisi1= np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1**3 #üzeri 3 yaptık
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g") # g yani green
```

```python
plt.plot(numpyDizisi2,numpyDizisi1,"g--") # g-- yani green çizgi çizgi
```

```python
plt.plot(numpyDizisi2,numpyDizisi1,"b*-") # b*- yani blue yıldız ve çizgi çizgi
```

```python
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1 sıra olacak 2 kolon olacak, şu an 2. grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure() #boş bir canvas oluşturduk
figureAxes = benimFigure.add_axes([0.2,0.2,0.9,0.9])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X ekseni Veri ismi")
figureAxes.set_ylabel("Y ekseni Veri ismi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()
eksen1 = figure2.add_axes([0.1,0.1,0.7,0.7])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3]) #0.2 ve 0.4 den başlayarak grafiğin duracağı yeri değiştirdik.Bu şekilde yerlerini değiştiriyoruz

eksen1.plot(numpyDizisi1,numpyDizisi2,"r")
eksen1.set_xlabel("X ekseni")
eksen1.set_ylabel("Y ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"y")
eksen2.set_xlabel("X ekseni")
eksen2.set_ylabel("Y ekseni")
eksen2.set_title("Küçük Grafik")
```

```python
plt.subplots()
```

```python
(benimFigure,benimEksenler) = plt.subplots(nrows=1,ncols=2)
type(benimEksenler)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")
plt.tight_layout()
```

```python
yeniFigur = plt.figure(dpi=100)

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])
yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label="numpyDizisi **2")
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label="numpyDizisi **3")
yeniEksen.legend(loc=6)
```

```python
yeniFigur.savefig("benimfigur.png",dpi=200)
```
